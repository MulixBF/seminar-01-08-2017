# Нахождение объектов ( object detection )

Другой важной задачей в обработке изображений является нахождение объетков на изображении. В самом простом виде она может быть сформулированна следующим образом - для каждого объекта, относящегося к известному классу, необходимо найти координаты наименьшего окна, полностью содержащего данный объект

# R-CNN

Одной из первых успешных моделей для нахождения объектов используя нейронные сети является [R-CNN](https://arxiv.org/pdf/1311.2524.pdf). В общих чертах алгоритм может быть описан следующим образом:

1. Генерация областей-кандидатов ( region proposal )
2. Для каждой области проводится процесс извлечения признаков с помощью конволюционной нейросетки
3. Для каждой области и для каждого класса проводится бинарная классификация с помощью SVM
4. Среди пересекающихся ( с некорым порогом по IOU ) регинов одного класса выбирается один с наибольшей оценкой

# Алгоритм выборочного поиска ( selective search )

В алгоритме R-CNN в общем виде может применяться любой метод определения областей-кандидатов. Однако, в оригинальной статье применялся алгоритм выборочного поиска:

1. На изображении выделяются небольшие сегменты на основании некоторой метрики похожести пикселей ([Felzebszwalb, Huttenlocher 2003](http://cs.brown.edu/~pff/papers/seg-ijcv.pdf))
2. Эти сегменты иерархически группируются жадным алгоритмом на основании некоторого набора метрик похожести, до тех пор пока не будет получен набор потенциальных объектов ([Uijlings et. al 2012](http://cs.brown.edu/~pff/papers/seg-ijcv.pdf))

Важно отметить, что данный алгоритм работает с пиксельными регионами, и ничего не знает про семантику.

# SPPNET - Spatial pyramid pooling networks

Следующим шагом в развитии алгоритмов нахождения объектов был SPPNET. Основной идеей этого алгоритма было использовать общую карту признаков для всего изображения, из которой извлекались признаки для отдельных регионов. Это позволило ускорить алгоритм на 1-2 порядка, за счёт того, что forward pass нейронной сети происходит один раз.

# Fast RCNN 

В 2012 году была предложена следующая доработка алгоритма - Fast RCNN. В ней, так же как и в SPPNET за один проход просчитывается карта признаков для всего изображения. После этого для каждого ROI ( region of interest ) производится операция max pooling специального вида - на выходе из неё получается вектор фиксированного размера ( 7x7 в случае VGG16 ). Далее используется полносвязная сетка с K+1 выходом, где K - количество классов. Дополнительный выход - это класс "background".

# Faster RCNN

В 2015 году был предложена очередная доработка данного класса алгоритмов - Faster RCNN.  Основным изменением по сравнению с Fast RCNN было внедрение специальной сети для генерации регионов кандидатов - RPN ( Region Proposal Network ). Эта модель учится предлагать качественные регионы кандидаты, при это используя общие с основной сетью конволюционные признаки. К этому моменту производительность основанной на VGG16 составила 5fps сети на современных GPU. Качество распознования при этом соответсвовало state-of-the-art уровню
